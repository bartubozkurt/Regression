```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt


# Deneyime göre maaş tahmini !!
```

```python
data = pd.read_csv('dataset/Salary_Data.csv')
```

```python
data.head()
```

```python
plt.plot(data.YearsExperience, data.Salary,'rx')
```

```python
X = data[['YearsExperience']]
y = data['Salary']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.3, random_state = 47)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linear_regressor = LinearRegression()

linear_regressor.fit(X_train, y_train)
```

```python
print(linear_regressor.intercept_) # kesişme b0 olacak

print(linear_regressor.coef_) # katsayı eğim b1 olacak
```

```python
y_pred = linear_regressor.predict(X_test)
```

```python
plt.plot(X_test,y_test,'rx')
plt.plot(X_test,y_pred,color = 'blue')
plt.show()
```

### Algoritma, maksimum test veri noktalarından geçen ve diğerlerinden minimum mesafeye sahip bir çizgi çizdi. Bu çizgi, "en uygun" veya regresyon çizgisi denir.


### Tecrübe arttıkça Maaş'ta artar POZİTİF artış.

## Doğrusal modelleri değerlendirmek için kullanılan üç temel ölçü vardır. Bunlar: Ortalama mutlak hata (MAE), Ortalama kare hatası (MSE) veya Kök ortalama kare hatası (RMSE).

```python
from sklearn import metrics

metrics.mean_absolute_error(y_test,y_pred)
```

```python
metrics.mean_squared_error(y_test, y_pred)
```

```python
np.sqrt(metrics.mean_squared_error(y_test, y_pred))
```
